# Visual Analysis

### Univariate Analysis
Univariate analysis focuses on analyzing one variable at a time to understand its distribution, central tendency, spread, and presence of outliers.

In this project, univariate analysis is used to understand the individual behavior of urban indicators before studying relationships between variables.

#### Variables Considered for Univariate Analysis

The following variables are analyzed individually:

- Happiness_Score (Target variable)
- Air_Quality_Index
- Cost_of_Living_Index
- Green_Space_Area
- Traffic_Density
- Healthcare_Index

These variables represent environmental, economic, and infrastructural aspects of urban life.


```python
# UNIVARIATE ANALYSIS
# Purpose: Understand distribution, spread, and frequency of individual variables

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle("Univariate Analysis of Urban Indicators", fontsize=16)

# 1. Happiness Score Distribution
sns.histplot(df['Happiness_Score'], kde=True, color='steelblue', ax=axes[0, 0])
axes[0, 0].set_title("Happiness Score Distribution")

# 2. Air Quality Index Distribution
sns.histplot(df['Air_Quality_Index'], kde=True, color='crimson', ax=axes[0, 1])
axes[0, 1].set_title("Air Quality Index Distribution")

# 3. Cost of Living Index Distribution
sns.histplot(df['Cost_of_Living_Index'], kde=True, color='darkorange', ax=axes[0, 2])
axes[0, 2].set_title("Cost of Living Index Distribution")

# 4. Green Space Area Spread (Outlier detection)
sns.boxplot(x=df['Green_Space_Area'], color='forestgreen', ax=axes[1, 0])
axes[1, 0].set_title("Green Space Area Spread")

# 5. Traffic Density Distribution (Categorical  FIXED)
sns.countplot(
    x='Traffic_Density',
    data=df,
    hue='Traffic_Density',
    palette='Set2',
    legend=True,
    ax=axes[1, 1]
)
axes[1, 1].set_title("Traffic Density Distribution")

# 6. Healthcare Index Distribution
sns.histplot(df['Healthcare_Index'], kde=True, color='purple', ax=axes[1, 2])
axes[1, 2].set_title("Healthcare Index Distribution")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

#### Key Observations from Univariate Analysis

- Happiness scores are mostly concentrated in the mid-to-high range, indicating moderate
  to good urban well-being across most cities.
- Air Quality Index shows a right-skewed distribution, suggesting the presence of a few
  highly polluted cities.
- Traffic density is predominantly medium to high, highlighting congestion as a common
  urban issue.
- Green space availability varies widely, with noticeable outliers.
- Cost of living shows moderate variation across cities.
- Healthcare index values are relatively consistent, indicating standardized access in
  most cities.


#### Traffic Density Distribution (Pie Chart)

A pie chart is used here to visualize the **proportional distribution** of traffic density
categories across cities.

Pie charts are suitable only for **low-cardinality categorical variables**, which makes
Traffic_Density an appropriate choice.


```python
# PIE PLOT
# Purpose: Show proportional distribution of Traffic Density categories

traffic_counts = df['Traffic_Density'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(
    traffic_counts,
    labels=traffic_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette('Set2')
)
plt.title("Traffic Density Proportion")
plt.axis('equal')
plt.show()
```

#### Conclusion  Univariate Analysis

Univariate analysis provided a clear understanding of the individual characteristics of
urban indicators. This step ensured that distributions, outliers, and category dominance
were identified before proceeding to bivariate and multivariate analysis.


## Bivariate Analysis

Bivariate analysis involves analyzing **two variables together** to understand their
relationship or comparison.

In this project, bivariate analysis is mainly used to study how different urban factors
affect the **Happiness_Score**, which is the target variable.


#### Variables Considered for Bivariate Analysis

The following comparisons are analyzed:

#### Numerical vs Numerical
- Air_Quality_Index vs Happiness_Score
- Green_Space_Area vs Happiness_Score
- Cost_of_Living_Index vs Happiness_Score

#### Categorical vs Numerical
- Traffic_Density vs Happiness_Score
- City vs Happiness_Score (Top 20 cities)

#### Categorical vs Categorical Analysis
- Year vs Traffic_Density
- Month Vs Traffic_Density


```python
# BIVARIATE ANALYSIS
# Purpose: Understand how individual factors influence Happiness Score

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle("Bivariate Analysis: Factors vs Happiness Score", fontsize=16)

# 1. Air Quality vs Happiness
sns.scatterplot(
    x='Air_Quality_Index',
    y='Happiness_Score',
    data=df,
    color='crimson',
    ax=axes[0, 0]
)
sns.regplot(
    x='Air_Quality_Index',
    y='Happiness_Score',
    data=df,
    scatter=False,
    ax=axes[0, 0]
)
axes[0, 0].set_title("Air Quality Index vs Happiness")

# 2. Green Space vs Happiness
sns.scatterplot(
    x='Green_Space_Area',
    y='Happiness_Score',
    data=df,
    color='forestgreen',
    ax=axes[0, 1]
)
sns.regplot(
    x='Green_Space_Area',
    y='Happiness_Score',
    data=df,
    scatter=False,
    ax=axes[0, 1]
)
axes[0, 1].set_title("Green Space Area vs Happiness")

# 3. Traffic Density vs Happiness (FIXED)
sns.boxplot(
    x='Traffic_Density',
    y='Happiness_Score',
    data=df,
    hue='Traffic_Density',
    palette='Set3',
    legend=True,
    ax=axes[1, 1]
)
axes[1, 1].set_title("Traffic Density vs Happiness")

# 4. Cost of Living vs Happiness
sns.scatterplot(
    x='Cost_of_Living_Index',
    y='Happiness_Score',
    data=df,
    color='darkorange',
    ax=axes[1, 0]
)
sns.regplot(
    x='Cost_of_Living_Index',
    y='Happiness_Score',
    data=df,
    scatter=False,
    ax=axes[1, 0]
)
axes[1, 0].set_title("Cost of Living Index vs Happiness")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
top_20_cities = (
    df.groupby('City', observed=True)['Happiness_Score']
      .mean()
      .sort_values(ascending=False)
      .head(20)
)

# Generate colors (one per bar)
colors = sns.color_palette("tab10", len(top_20_cities))

# Plot using matplotlib (stable)
plt.figure(figsize=(14, 5))
plt.bar(
    top_20_cities.index,
    top_20_cities.values,
    color=colors
)

plt.title("Top 20 Cities by Average Happiness Score")
plt.xlabel("City")
plt.ylabel("Average Happiness Score")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Plot 1: Traffic Density vs Year
sns.countplot(
    x='Year',
    hue='Traffic_Density',
    data=df,
    palette='Set1',
    ax=axes[0]
)
axes[0].set_title("Traffic Density Across Years")
axes[0].set_xlabel("Year")
axes[0].set_ylabel("Count")

# Plot 2: Traffic Density vs Month
sns.countplot(
    x='Month',
    hue='Traffic_Density',
    data=df,
    palette='Set2',
    ax=axes[1]
)
axes[1].set_title("Traffic Density Across Months")
axes[1].set_xlabel("Month")
axes[1].set_ylabel("Count")
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

### Key Observations from Bivariate Analysis

- Higher air pollution levels are generally associated with lower happiness scores.
- Cities with more green space tend to show higher happiness levels.
- Higher traffic density negatively impacts happiness.
- Cost of living shows a weak negative relationship with happiness, indicating that
  higher costs do not necessarily translate to better quality of life.


### Conclusion  Bivariate Analysis

Bivariate analysis revealed how individual environmental and economic factors directly
influence urban happiness. This step helped identify the most impactful variables before
examining combined effects using multivariate analysis.


## Multivariate Analysis

Multivariate analysis involves analyzing **three or more variables simultaneously**
to understand combined effects and interactions.

This analysis helps uncover complex relationships that are not visible in univariate
or bivariate analysis.


### Variables Considered for Multivariate Analysis

- Happiness_Score
- Air_Quality_Index
- Green_Space_Area
- Traffic_Density
- Cost_of_Living_Index
- Healthcare_Index
- Year

These variables together represent environmental, economic, and temporal dimensions
of urban well-being.


```python
# MULTIVARIATE ANALYSIS  CORRELATION HEATMAP
# Purpose: Identify strength and direction of relationships among numerical variables

numeric_cols = [
    'Decibel_Level',
    'Green_Space_Area',
    'Air_Quality_Index',
    'Happiness_Score',
    'Cost_of_Living_Index',
    'Healthcare_Index',
    'Year'
]

plt.figure(figsize=(10, 8))
sns.heatmap(
    df[numeric_cols].corr(),
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)
plt.title("Correlation Heatmap of Urban Indicators")
plt.show()
```

#### Interpretation of Correlation Heatmap

- Happiness score shows strong relationships with environmental factors such as air quality
  and green space.
- Traffic-related and pollution indicators negatively correlate with happiness.
- Healthcare and economic indicators show moderate influence.


```python
# MULTIVARIATE ANALYSIS  AQI vs Happiness with Traffic Density
# Purpose: Understand how traffic density modifies the impact of air pollution on happiness

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Air_Quality_Index',
    y='Happiness_Score',
    hue='Traffic_Density',
    data=df,
    palette='Set1'
)
plt.title("Air Quality vs Happiness by Traffic Density")
plt.show()
```

```python
# MULTIVARIATE ANALYSIS  Traffic Density vs Happiness Across Years
# Purpose: Compare present (2024) and future (2030) trends

plt.figure(figsize=(10, 6))
sns.boxplot(
    x='Traffic_Density',
    y='Happiness_Score',
    hue='Year',
    data=df,
    palette='coolwarm'
)
plt.title("Traffic Density vs Happiness Across Years")
plt.show()
```

#### Conclusion  Multivariate Analysis

Multivariate analysis demonstrated that urban happiness is influenced by a combination
of environmental quality, traffic congestion, economic conditions, and time.
Balanced urban planning is therefore essential to improve overall well-being.


```python
# FINAL OUTCOME: Top Cities to Live In
# Criteria: High Happiness Score and High Healthcare Index

top_cities_happy_health = (
    df.groupby('City', observed=True)
      .agg({
          'Happiness_Score': 'mean',
          'Healthcare_Index': 'mean'
      })
      .sort_values(
          by=['Happiness_Score', 'Healthcare_Index'],
          ascending=[False, False]
      )
)

# Display top 10 cities
top_cities_happy_health.head(10)
```

## Final Summary and Conclusion

This project performed a comprehensive Exploratory Data Analysis (EDA) on the **Urban Bliss Index (20242030)** dataset to understand the factors influencing urban well-being and livability across cities.

### Key Analytical Steps
- **Data preparation** ensured clean, consistent, and duplicate-free records suitable for analysis.
- **Univariate analysis** was conducted to understand the distribution and variability of individual indicators such as happiness score, healthcare index, cost of living, air quality, green space, noise levels, and traffic density.
- **Bivariate analysis (visual and non-visual)** examined relationships between:
  - Environmental factors (air quality, noise, green space) and happiness.
  - Traffic density and happiness levels.
  - Healthcare quality and overall well-being.
- **City-level analysis** focused on the top-performing cities to improve interpretability while handling high cardinality.
- **Temporal analysis** compared patterns across years and months to identify changes and seasonality.

### Key Insights
- Environmental factors such as **air quality, noise levels, and green space availability** show a strong association with urban happiness.
- **Traffic congestion** is consistently linked with lower happiness scores.
- **Healthcare quality** plays a significant role in overall well-being and long-term livability.
- Higher **cost of living** does not necessarily guarantee higher happiness, indicating diminishing returns of economic factors alone.
- Certain cities consistently rank higher due to **balanced performance across multiple indicators**, rather than excelling in a single metric.

### Final Outcome
Cities with **high happiness scores and strong healthcare infrastructure** emerge as the most desirable places to live. The analysis highlights that sustainable urban development requires a balanced approach combining environmental quality, infrastructure, and social well-being.

### Conclusion
The Urban Bliss Index dataset demonstrates that **urban happiness is multidimensional**. Effective urban planning should prioritize environmental sustainability, healthcare access, and congestion management alongside economic growth to enhance overall quality of life.
